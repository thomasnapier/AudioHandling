```python
def purity_score(y_true, y_pred):
    # compute contingency matrix (also called confusion matrix)
    confusion_matrix = contingency_matrix(y_true, y_pred)
    # return purity
    return np.sum(np.amax(confusion_matrix, axis=0)) / np.sum(confusion_matrix)
```

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Assuming you have your data stored in a variable called "data"
# Make sure your data is in the shape of (n_samples, n_features)

# Perform PCA dimension reduction
pca = PCA()
pca.fit(feature_vectors)

# Calculate the cumulative explained variance ratio
explained_variance_ratio = np.cumsum(pca.explained_variance_ratio_)

# Plot the explained variance ratio
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('Explained Variance Ratio by Number of Principal Components')
plt.show()

# Get the weights of the principal components
weights = np.abs(pca.components_)

# Plot the weights of the principal components
plt.figure(figsize=(8, 6))
for i in range(len(weights)):
    plt.plot(range(1, len(weights[i]) + 1), weights[i], label=f'PC {i+1}')
plt.xlabel('Features')
plt.ylabel('Weights')
plt.title('Weights of Principal Components')
plt.legend()
plt.show()
```

```python
# Assuming you have already computed the weights of the principal components
weights = np.abs(pca.components_)

# Calculate the sum of absolute weights for each principal component
sum_of_weights = np.sum(weights, axis=1)

# Find the principal component with the highest sum of weights
max_weight_idx = np.argmax(sum_of_weights)

# Print the index and corresponding sum of weights
print("Principal Component with Highest Weighting:")
print(f"Index: PC {max_weight_idx + 1}")
print(f"Sum of Weights: {sum_of_weights[max_weight_idx]}")
```

```python
# Assuming you have already computed the weights of the principal components
weights = np.abs(pca.components_)

# Find the indices of the two principal components with the highest weights
top_two_indices = np.argsort(-np.sum(weights, axis=1))[:2]

# Select the corresponding top two principal components
top_two_components = pca.components_[top_two_indices]
top_two_weights = weights[top_two_indices]

# Project the data onto the top two principal components
projected_data = np.dot(feature_vectors, top_two_components.T)

# Create colormap based on labels
label_colors = labels.astype(int)  # Convert labels to integers
cmap = plt.cm.get_cmap('viridis')  # Replace with the desired colormap

# Plot the reduced dimensions in a 2D scatterplot with colormap
plt.scatter(projected_data[:, 0], projected_data[:, 1], c=label_colors, cmap=cmap, s=10)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Data Projection onto Top Two Principal Components')
plt.show()

# Print the weightings of the top two principal components
print("Weightings of the Top Two Principal Components:")
for i in range(len(top_two_indices)):
    print(f"Principal Component {i+1}:")
    for j in range(len(top_two_weights[i])):
        print(f"Feature {j+1}: {top_two_weights[i][j]}")
    print()
```
