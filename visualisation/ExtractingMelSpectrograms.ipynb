```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
```

Loading audio files

```python
geo_path = "D:\\PhD\\data\\test\\geophony"
anthro_path = "D:\\PhD\\data\\test\\anthrophony"
bio_path = "D:\\PhD\\data\\test\\biophony"
other_path = "D:\\PhD\\data\\test\\other"
output_path = "D:\\PhD\\data\\test"
audio_path = "D:\\PhD\\data\\ESC-50-master\\audio"


file_paths = [geo_path, anthro_path, bio_path, other_path]
```

Code for iterating through csv files to obtain and copy audio files

```python
import os
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
import numpy as np
import pandas as pd
import csv
import shutil

#for each csv file, for each line get file name and append to audio path
for path in file_paths:
    for subdir, dirs, files in os.walk(path):
        for file in files:
            filepath = subdir + os.sep + file
            print(filepath)

            # iterate through csv
            with open(filepath, 'r') as file:
                reader = csv.reader(file)
                next(reader, None) #skip header
                for row in reader:
                    print(row[0])
                    audio_file = row[0]
                    shutil.copyfile(audio_path + "\\" + audio_file, path + "\\" + audio_file)

            
```

Code for iterating through folders of audio clips and converting them to mel spectrograms

```python
import os
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
import numpy as np
import pandas as pd

for path in file_paths:
    for subdir, dirs, files in os.walk(path):
        for file in files:
            filepath = subdir + os.sep + file

            if filepath.endswith(".wav"):
                # load audio files with librosa
                y, sr = librosa.load(filepath)
                y = y[:100000] # shorten audio a bit for speed
                window_size = 1024
                window = np.hanning(window_size)
                stft  = librosa.core.spectrum.stft(y, n_fft=window_size, hop_length=512, window=window)
                out = 2 * np.abs(stft) / np.sum(window)

                # For plotting headlessly
                from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas

                fig = plt.Figure()
                canvas = FigureCanvas(fig)
                ax = fig.add_subplot(111)
                p = librosa.display.specshow(librosa.amplitude_to_db(out, ref=np.max), ax=ax, y_axis='log', x_axis='time')

                fig.savefig("{}\\{}.png".format(path, file[:-4]))


            
```
